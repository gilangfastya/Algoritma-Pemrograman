```python
print("LATIHAN 1")

# List awal
buah = ['apel', 'mangga', 'jeruk', 'anggur', 'pisang']
print(f"List awal\t\t\t\t: {buah}")

# a. Tambahkan 'semangka' di akhir list
buah.append('semangka')
print(f"Setelah menambah 'semangka'\t\t: {buah}")

# b. Sisipkan 'durian' di antara 'jeruk' dan 'anggur'
buah.insert(4, 'durian')  # index 4 setelah 'jeruk' dan sebelum 'anggur'
print(f"Setelah menyisipkan 'durian'\t\t: {buah}")

# c. Hapus 'mangga' dari list
buah.remove('mangga')
print(f"Setelah menghapus 'mangga'\t\t: {buah}")

# d. Ubah 'pisang' menjadi 'nanas'
index_pisang = buah.index('pisang')
buah[index_pisang] = 'nanas'
print(f"Setelah mengubah 'pisang' jadi 'nanas'\t: {buah}")

# e. Tampilkan 3 buah pertama
print(f"Tiga buah pertama\t\t\t: {buah[:3]}")

print("\nLATIHAN 2")

# List awal
angka = [45, 12, 78, 23, 56, 89, 34]
print(f"List awal\t\t\t\t: {angka}")

# a. Urutkan list secara ascending
angka.sort()
print(f"Urutan ascending\t\t\t: {angka}")

# b. Urutkan list secara descending
angka.sort(reverse=True)
print(f"Urutan descending\t\t\t: {angka}")
```
