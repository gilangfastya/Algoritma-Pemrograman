```python
import datetime
import os
import string
import random

buku_template = {
    'judul': 'Judul Buku',
    'penulis': 'Penulis Buku',
    'tahun_terbit': 0000,
    'tgl_terakhir_pinjam': datetime.datetime(1111, 1, 1),
    'status': 'TERSEDIA'
}

data_buku = {}

while True:
    os.system("cls")

    print(f"{'SISTEM MANAJEMEN PERPUSTAKAAN':^50}")
    print("-"*60)

    buku = dict.fromkeys(buku_template.keys())
    buku['judul'] = input("Judul Buku: ")
    buku['penulis'] = input("Penulis Buku: ")
    buku['tahun_terbit'] = int(input("Tahun Terbit: "))

    print("Masukkan tanggal terakhir pinjam:")
    tahun = int(input("Tahun (YYYY): "))
    bulan = int(input("Bulan (1-12): "))
    tanggal = int(input("Tanggal (1-31): "))
    buku['tgl_terakhir_pinjam'] = datetime.datetime(tahun, bulan, tanggal)

    buku['status'] = input("Status (TERSEDIA/TIDAK): ").upper()

    KEY = ''.join((random.choice(string.ascii_uppercase) for i in range(6)))
    data_buku.update({KEY: buku})

    print(f"\n{'KEY':<8} {'JUDUL':<25} {'PENULIS':<20} {'TAHUN':<10} {'TERAKHIR PINJAM':<18} {'STATUS':<10}")
    print("-"*90)
    for key in data_buku:
        JUDUL = data_buku[key]['judul']
        PENULIS = data_buku[key]['penulis']
        TAHUN = data_buku[key]['tahun_terbit']
        TGL_PINJAM = data_buku[key]['tgl_terakhir_pinjam'].strftime("%x")
        STATUS = data_buku[key]['status']
        print(f"{key:<8} {JUDUL:<25} {PENULIS:<20} {TAHUN:<10} {TGL_PINJAM:<18} {STATUS:<10}")

    print("\n")
    ulang = input("Tambah buku lagi (y/n)? ").lower()
    if ulang == 'n':
        break

print("\nAkhir dari program, terima kasih.")
```
