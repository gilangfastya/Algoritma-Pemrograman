```python
# latihan soal nomor 1

# a. buat tuple
mahasiswa = ("Gilang Wijaya", "12345678", "Teknik Informatika")
print("mahasiswa (awal) =", mahasiswa)

# b. ubah prodi -> ubah ke list dulu baru ke tuple lagi
mahasiswa_list = list(mahasiswa)
mahasiswa_list[2] = "Teknologi Informasi dan Elektro"
mahasiswa = tuple(mahasiswa_list)
print("mahasiswa (ubah) =", mahasiswa)

# c. tampilkan NIM dan Prodi
nim = mahasiswa[1]
prodi = mahasiswa[2]
print("NIM \t=", nim)
print("Prodi \t=", prodi)
```

```python
# latihan soal nomor 2

keranjang = {"apel": 5000, "pisang": 3000, "jeruk": 7000, "anggur": 15000}

total = 0
for buah, harga in keranjang.items():
    print(f"Harga {buah}\t: Rp {harga}")
    total += harga

print("Total belanja\t: Rp", total)
```
