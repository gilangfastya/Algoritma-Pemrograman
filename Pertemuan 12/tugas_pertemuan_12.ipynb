```python
# luas persegi panjang
def luas_persegi_panjang(panjang, lebar):
  return panjang * lebar

p = float(input("masukkan panjang\t: "))
l = float(input("masukkan lebar\t\t: "))

hasil = luas_persegi_panjang(p,l)
print(f"luas persegi panjang\t= {hasil:.2f}")
```

```python
# volume persegi panjang
def volume_persegi_panjang(panjang, lebar, tinggi):
  return panjang * lebar * tinggi

p = float(input("masukkan panjang\t: "))
l = float(input("masukkan lebar\t\t: "))
t = float(input("masukkan tinggi\t\t: "))

hasil = volume_persegi_panjang(p,l,t)
print(f"volume persegi panjang\t= {hasil:.2f}")
```

```python
# luas lingkaran
def luas_lingkaran(jari_jari):
  return math.pi * (jari_jari ** 2)

r = float(input("masukan bilangan\t: "))
hasil = luas_lingkaran(r)
print(f"luas lingkaran\t\t= {hasil:.2f}")
```

```python
# volume bola
def volume_bola(r):
  return (4/3) * math.pi * (r ** 3)

r = float(input("masukan bilangan\t: "))
hasil = volume_bola(r)
print(f"volume bola\t\t= {hasil:.2f}")
```
